```python
# Analysing Difference between Levels (Resampling Weekly)
#MOVING AVERAGES
#person had rs 10 at start of the day and at end of day he earns rupees 50 
#another person: start of day has 100 and at end of day earns 120
#who earned more and in what percentage
#50 - 10 = 40
#120 - 100 = 20
#first person got greater share than person 2
#same thing for companies, we need to see open and close prices

ibm['diff'] = ibm['Open'] - ibm['Close'] 
ibm_diff = ibm.resample('W').mean() # resampling weekly
ibm_diff.tail(10)

#from diff col, we see ibm, stocks did pretty well, then they got low, -1.4
#then they were little higher, -0.04
#then they started making profit and so on
```

```python
# Step 3: Moving Windows
# Moving Windows are there when you compute the statistic on a window of data
#represented by a particular period of time and then slide the window across the data by a specified interval.

#that way, statistic is continually calculated as long as the window falls first with in the dates of the T.S

#A rolling mean smoothens out short term flucuations and highlight longer term trends in data

#Inshort: your starting from home at slower speed, then you gradually increase your speed
#thats how moving windows work

# Daily Percentages
#suppose i take Adj Close 15 - 09 - 19 - Adj Close 14 - 09 - 19 % (daily percentages)

daily_close_ibm = ibm[['Adj Close']]

# Daily returns
daily_pct_change_ibm = daily_close_ibm.pct_change()

# Replace NA values with 0
daily_pct_change_ibm.fillna(0, inplace=True)

daily_pct_change_ibm.head()
```

```python
daily_pct_change_ibm.hist(bins=50)

# Show the plot
plt.show()
#the stocks have not made profits more than 0.05 and havent lost more than -0.075
```

```python
#we do this for NVDA and other companies
daily_close_nvda = nvda[['Adj Close']]

# Daily returns
daily_pct_change_nvda = daily_close_nvda.pct_change()

# Replace NA values with 0
daily_pct_change_nvda.fillna(0, inplace=True)

daily_close_intc = intc[['Adj Close']]

# Daily returns
daily_pct_change_intc = daily_close_intc.pct_change()

# Replace NA values with 0
daily_pct_change_intc.fillna(0, inplace=True)

daily_close_qcom = qcom[['Adj Close']]

# Daily returns
daily_pct_change_qcom = daily_close_qcom.pct_change()

# Replace NA values with 0
daily_pct_change_qcom.fillna(0, inplace=True)

daily_close_amd = amd[['Adj Close']]

# Daily returns
daily_pct_change_amd = daily_close_amd.pct_change()

# Replace NA values with 0
daily_pct_change_amd.fillna(0, inplace=True)

daily_pct_change_amd.head()
```

```python
# Step 4: Volatality of Stocks in terms of variance
#The volatility of stock is a measurement of the change in variance in the returns of a stock over a specific
#period of time

#To compare the volatility of stock with another stock to get a feel for which may have less risk 
# or to a market index to examine the stock's volatility in the overall market

#Generally the higher the volatility, the riskier the investment in that stock

#if its volatility, you can risk more or gain more
import numpy as np
```

```python
min_periods = 75 

# Calculate the volatility
vol = daily_pct_change_ibm.rolling(min_periods).std() * np.sqrt(min_periods) 
#i am getting variance of daily percentage values

vol.fillna(0,inplace=True)

vol.tail()
#we see ibm stocks are less volatile
```

```python
# Plot the volatility
vol.plot(figsize=(10, 8))

# Show the plot
plt.show()
```

```python
# Rolling Means (Trends and Seasonality), this is to see for long term
#Variance gives you a better picture of volatility

ibm_adj_close_px = ibm['Adj Close']
# Short moving window rolling mean
ibm['42'] = ibm_adj_close_px.rolling(window=40).mean()
#here i am getting short term window
#40 will give me quaterly trend

# Long moving window rolling mean, long term window, this will give yearly trend
ibm['252'] = ibm_adj_close_px.rolling(window=252).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
ibm[['Adj Close', '42', '252']].plot(title="IBM")

# Show plot
plt.show()

#same thing for ibm, i have done for nvda
nvda_adj_close_px = nvda['Adj Close']
# Short moving window rolling mean
nvda['42'] = nvda_adj_close_px.rolling(window=40).mean()

# Long moving window rolling mean
nvda['252'] = nvda_adj_close_px.rolling(window=252).mean()

# Plot the adjusted closing price, the short and long windows of rolling means
nvda[['Adj Close', '42', '252']].plot(title="NVIDIA")

# Show plot
plt.show()

#42 is the shorter window, if you have to invest in monthly or quarterly basis, invest in this stock
#252: if you want to invest in long term, then invest in this
#ibm was high in 2012 and 2013 and then a fall
#you can invest for long term in ibm for 1 or 2 years, as they are stable
#you can see trend and seasonality
#short window shows the trend
#longer window shows the seasonality
```
